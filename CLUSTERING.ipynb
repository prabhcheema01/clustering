```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, PowerTransformer, FunctionTransformer, MaxAbsScaler
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from yellowbrick.cluster import KElbowVisualizer
import matplotlib.pyplot as plt
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv"
data = pd.read_csv(url)
data.head()
```

```python
X = data.drop(columns=['Region', 'Channel'])
X.head()
```

```python
y_dummy = np.zeros(X.shape[0])
preprocessing_techniques = {
    "Normalization (MinMaxScaler)": MinMaxScaler(),
    "Standardization (StandardScaler)": StandardScaler(),
    "Robust Scaling": RobustScaler(),
    "PCA": PCA(n_components=2),
    "Feature Selection (SelectKBest)": SelectKBest(score_func=f_classif, k=2),
    "Log Transformation": FunctionTransformer(np.log1p, validate=True),
    "Box-Cox Transformation": PowerTransformer(method='box-cox', standardize=True),
    "Imputation": SimpleImputer(strategy='mean'),
    "Feature Scaling with MaxAbsScaler": MaxAbsScaler()
}
```

```python
evaluation_metrics = {
    "Silhouette Score": silhouette_score,
    "Calinski-Harabasz Index": calinski_harabasz_score,
    "Davies-Bouldin Index": davies_bouldin_score
}
```

```python
clustering_algorithms = {
    "KMeans": KMeans,
    "Agglomerative Clustering": AgglomerativeClustering,
    "DBSCAN": DBSCAN
}
```

```python
results = []

for n_clusters in [3, 4, 5]:
    for prep_name, prep_method in preprocessing_techniques.items():
        for cluster_name, cluster_algo in clustering_algorithms.items():
            preprocessor = ColumnTransformer(
                transformers=[('preprocessor', prep_method, X.columns)]
            )

            if cluster_name == 'KMeans':
                pipeline = Pipeline([
                    ('preprocessor', preprocessor),
                    ('clf', cluster_algo(n_clusters=n_clusters))
                ])

                plt.figure(figsize=(8, 6))
                visualizer = KElbowVisualizer(pipeline.named_steps['clf'], k=(2, 6))
                visualizer.fit(X)
                optimal_k = visualizer.elbow_value_
                visualizer.show()
                plt.tight_layout()
                plt.show()

            pipeline = Pipeline([
                ('preprocessor', preprocessor),
                ('clf', cluster_algo(n_clusters=n_clusters) if cluster_name != 'DBSCAN' else cluster_algo())
            ])

            pipeline.fit(X, y_dummy)
            labels = pipeline.named_steps['clf'].labels_
            if len(np.unique(labels)) <= 1:
                print(f"Skipping evaluation for {cluster_name} with {prep_name} preprocessing technique due to single cluster detection")
                continue

            scores = {}
            for metric_name, metric_func in evaluation_metrics.items():
                score = metric_func(X, labels)
                scores[metric_name] = score

            results.append({"Preprocessing": prep_name, "Algorithm": cluster_name, "Clusters": n_clusters, "Optimal K": None if cluster_name == 'DBSCAN' else optimal_k, **scores})

```

```python
results_df = pd.DataFrame(results)
print(results_df)
```
